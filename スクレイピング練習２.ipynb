```python
import requests

from bs4 import BeautifulSoup
```

```python
page_url = "https://imagesearch.excite.co.jp/?q=%E3%83%95%E3%82%A9%E3%83%BC%E3%82%AF%E3%83%AA%E3%83%95%E3%83%88&page=1"

r = requests.get(page_url)

soup = BeautifulSoup(r.text, features="html.parser")
```

```python
soup.find_all("img", attrs = {"alt": "フォークリフト"})
```

```python
img_tags = soup.find_all("img", attrs = {"class": "lazyload"})
```

```python
img_urls = []

for img_tag in img_tags:
    url = img_tag.get('src')

    img_urls.append(url)

    print(url)
```

## ダウンロードイメージ関数

```python
def download_image(url, file_path):
    r = requests.get(url, stream=True) #後で調べたほうがいいよ

    if r.status_code == 200:
        with open(file_path, "wb") as f:
            f.write(r.content)
```
