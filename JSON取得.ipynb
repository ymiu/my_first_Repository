```
# 天気予報を取得するモジュール
# URLからJSONデータをダウンロードする関数
def get_json(url):
    import urllib.request as req
    import json
    #データをダウンロード
    res = req.urlopen(url)
    json_data = res.read()
    # JSONデータをPythonで扱えるよう読み込む
    return json.loads(json_data)

# 都市IDを取得することで天気情報を取得する関数
def get_weather(city_id):
    url = 'https://weather.tsukumijima.net/api/forecast'
    url += '?city='+str(city_id)
    data = get_json(url)
    s = ''
    
    for row in data['forecasts']: # dataの「forecasts」を1行づつ読み込む
        label = row['dateLabel'] # 読み込んだ行の「datalabel」の値を代入
        telop = row['telop'] # 読み込んだ行の「telop」の値を代入
        s += label + ' :' + telop + '\n'
    return s

#　モジュールとして利用されるか判定
if __name__ =='__main__':
    res = get_weather(410020) # 東京の天気を取得 
    print(res)
```
